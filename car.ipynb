```python
from operator import length_hint
import pandas as pd

def predict(obj):

  if obj['safety'] == 'low':
    return 'unacc'

  if obj['buying'] == 'high':
    if obj['safety'] == 'med' or 'high':
      return 'acc'

  elif obj['buying'] != 'vhigh' or 'high':
    if obj['safety'] == 'med' or 'low':
      return 'unacc'

  elif obj['buying'] == 'low' or 'med':
    if obj['maint'] == 'low':
      if obj['doors'] == '4':
        if obj['lug_boot'] == 'med' or 'big':
          if obj['safety'] == 'high':
            return 'vgood'

  elif obj['buying'] == 'med':
     if obj['maint'] == 'med':
      if obj['doors'] == '3' or '4':
        if obj['lug_boot'] == 'med':
          if obj['safety'] == 'med':
            return 'good'


data = pd.read_csv('car.data', header=None, names=['buying','maint','doors','persons','lug_boot','safety','class'])

count = 0
for i , obj in data.iterrows():
  print(obj['class'] ,'-', predict(obj))

  #Calcular acurácia

  if obj['class'] == predict(obj):
    count +=1

print(count/len(data))








```
